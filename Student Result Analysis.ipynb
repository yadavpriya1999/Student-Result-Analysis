```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Dataset loading
df=pd.read_csv("students_score.csv")
print(df.head())
```

```python
#Drop column
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Drop Unmaned Column

```python
df=df.drop("Unnamed: 0",axis=1)
```

```python
df.head()
```

# Gender Distribution Analysis

```python
##Figure Size-Pass height& width
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender")

#For data Labeling
ax.bar_label(ax.containers[0])
plt.show()
```

From the Above Char we have analysed that:
the number of female is more than the number of male

# Parents Education Impact Analysis

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
```

```python
print(gb)
```

```python
#heat map
plt.figure(figsize=(5,5))
sns.heatmap(gb,annot=True)
plt.show()
```

# Parents Maritalstatus Impact Analysis

```python
gb1=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
```

```python
print(gb1)
```

```python
#heat map
plt.figure(figsize=(7,5))
sns.heatmap(gb1,annot=True)
plt.show()
```

From the above chart we have concalued that  there is on/negligible impact on the student's score due to their parents marital status

# Distribution of Ethnic Groups

```python
#Percentage Distribution-Py chart
groupA=df.loc[(df['EthnicGroup']=="group A")].count()
groupB=df.loc[(df['EthnicGroup']=="group B")].count()
groupC=df.loc[(df['EthnicGroup']=="group C")].count()
groupD=df.loc[(df['EthnicGroup']=="group D")].count()
groupE=df.loc[(df['EthnicGroup']=="group E")].count()

#Pi chart 

#Labels
l=["group A","group B","group C","group D","group E"]
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
print(mlist)
plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
